```python
def max_of_all(a,b,c):
    if(a>b ):
        if(c>a):
            print(f"{c} is the max ")
        else:
            print(f"{a} is the max ")
    elif(b>a):
        if(c>b):
            print(f"{c} is the max ")
        else:
            print(f"{b}is the max")    

max_of_all(2,5,9)
```

```python
def count_vowels(s):
    count = 0
    for x in s:
        if x in (("a","e","i","o","u")):
            count+=1
    print(count)
count_vowels("Hello World")
```

```python
def recursion(n):
    fact =1
    while n>=1:
        fact = fact *n
        n-=1

    print(fact)

    
recursion(6)
    
```
