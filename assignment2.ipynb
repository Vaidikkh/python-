1. Find the largest, smallest, and second largest number from 10 user
inputs.


```python

numbers = []

for i in range(10):
    num = int(input("Enter a number: "))
    numbers.append(num)

numbers.sort()

smallest = numbers[0]
largest = numbers[-1]
second_largest = numbers[-2]

print("Largest Number =", largest)
print("Smallest Number =", smallest)
print("Second Largest Number =", second_largest)

```

# 2.  Count vowels, consonants, digits, and special characters in a string. 


```python
string = input("Enter a string: ")

vowels = 0
consonants = 0
digits = 0
special = 0

for ch in string:
    if ch.isalpha():
        if ch.lower() in "aeiou":
            vowels += 1
        else:
            consonants += 1
    elif ch.isdigit():
        digits += 1
    elif not ch.isspace():
        special += 1

print("Vowels =", vowels)
print("Consonants =", consonants)
print("Digits =", digits)
print("Special Characters =", special)
```

# 3. Check whether a number is an Armstrong number.

```python
num = int(input("Enter a number: "))

temp = num
sum = 0
n = len(str(num))

while temp > 0:
    digit = temp % 10
    sum += digit ** n
    temp //= 10

if sum == num:
    print(num, "is an Armstrong Number")
else:
    print(num, "is not an Armstrong Number")
```

# 4. Find duplicate elements and their frequency in a tuple.

```python
t = (1, 2, 3, 2, 4, 5, 1, 2, 3, 6)

print("Tuple:", t)

for item in set(t):
    count = t.count(item)
    if count > 1:
        print(item, "appears", count, "times")
```

5. Find the topper from a tuple of student names and marks. 

```python
students = (
    ("Vaidik", 95),
    ("uv", 92),
    ("vbh", 88),
    ("plk", 95),
    ("Pawan", 90)
)

topper = students[0]

for student in students:
    if student[1] > topper[1]:
        topper = student

print("Topper Name:", topper[0])
print("Marks:", topper[1])
```

# 6. Count the frequency of each word in a sentence using a dictionary.



```python
sentence = input("Enter a sentence: ")

words = sentence.split()
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

print("Word Frequencies:")
for word, count in frequency.items():
    print(word, ":", count)
```

7. Display students scoring above the average marks from a dictionary

Check whether a triangle is Equilateral, Isosceles, or Scalene. 

```python
a = float(input("Enter first side: "))
b = float(input("Enter second side: "))
c = float(input("Enter third side: "))

if a == b == c:
    print("Equilateral Triangle")
elif a == b or b == c or a == c:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle") 
```
