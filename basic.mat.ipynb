```python
point  =np.array([1,2,3,4])
plt.pie(point)
plt.show()
```

```python
point  =np.array([1,2,3,4])
label = np.array(["Co2","N2","O2","other"])

plt.pie(point,labels =label)
plt.show()
```

```python
point  =np.array([3,6,2,0])
label = np.array(["Co2","N2","O2","other"])
color = np.array(["red","green","magenta","blue"])

plt.pie(point,labels =label,colors = color)
plt.show()
```

```python
point  =np.array([3,6,1.1,0.9])
label = np.array(["Co2","N2","O2","other"])
color = np.array(["red","green","magenta","blue"])

plt.pie(point,labels =label,colors = color,shadow =True)
plt.show()
```

```python
point  =np.array([3,6,1.1,0.9])
label = np.array(["Co2","N2","O2","other"])
color = np.array(["red","green","magenta","blue"])
x = np.array([0.0,0.2,0.1,0.0])
plt.pie(point,labels =label,colors = color,shadow =True,explode =x)
plt.show()
```

```python
point  =np.array([3,6,1.1,0.9])
label = np.array(["Co2","N2","O2","other"])
color = np.array(["red","green","magenta","blue"])
x = np.array([0.0,0.2,0.1,0.0])
plt.pie(point,labels =label,colors = color,shadow =True,explode =x)
plt.legend()
plt.show()
```

```python
point  =np.array([3,6,1.1,0.9])
label = np.array(["Co2","N2","O2","other"])
color = np.array(["red","green","magenta","blue"])
x = np.array([0.0,0.2,0.1,0.0])
plt.pie(point,labels =label,colors = color,shadow =True,explode =x)
plt.legend('mypiechart')
plt.show()
```
