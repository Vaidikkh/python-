4. Create a DataFrame using the dataset below:
Name Age Salary
A 25 30000
B NaN 40000
C 27 NaN
D 24 35000
a) Tasks:
● Identify missing values
● Fill missing Age with average age
● Fill missing Salary with median salary
● Display cleaned dataset
b) Expected Concepts:
● isnull()
● fillna()
● mean()
● median()

isnull(): It  checks for missing  values in a DataFrame
fillna(): It replaces missing values with a  value
mean(): it calculates the average values in a column
median(): it returns the middle value of the column

```python
import pandas as pd
import numpy as np

data = {
    "Name":["A","B","C","D"],
    "Age":[25,np.nan,27,24],
    "Salary":[30000,40000,np.nan,35000]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.isnull())
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
print(df)
```
