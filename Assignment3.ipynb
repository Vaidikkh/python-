Write a program that takes a list of numbers and prints all prime numbers in the list using loops. Do not use any built-in prime-checking functions.

```python
   numbers = [2, 4, 5, 8, 11, 15, 17, 20]

print("Prime Numbers:")

for num in numbers:
    if num > 1:
        is_prime = True

        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            print(num)
```

Write a user-defined function that accepts a list of integers and returns:
• Count of even numbers
• Sum of odd numbers

```python
def calculate(numbers):
    even_count = 0
    odd_sum = 0

    for num in numbers:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_sum += num

    return even_count, odd_sum


nums = [1, 2, 3, 4, 5, 6]

even_count, odd_sum = calculate(nums)

print("Count of Even Numbers:", even_count)
print("Sum of Odd Numbers:", odd_sum)
```

Create a function that calculates simple interest. Use default arguments for rate and time. Demonstrate calling the function using both positional and keyword arguments.

```python
def simple_interest(principal, rate=5, time=2):
    si = (principal * rate * time) / 100
    return si


# Positional Arguments
print("Simple Interest:", simple_interest(10000))

# Keyword Arguments
print("Simple Interest:", simple_interest(principal=10000, rate=8, time=3))
```

Create a BankAccount class with private attributes for balance. Provide methods to deposit and withdraw money. Ensure balance cannot be accessed directly.

```python
class BankAccount:

    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def show_balance(self):
        print("Current Balance:", self.__balance)


account = BankAccount(5000)

account.deposit(2000)
account.withdraw(1500)
account.show_balance()
```

Write a program that takes two numbers from the user and performs division. Handle Division by zero and Invalid input type.

```python
try:
    num1 = float(input("Enter First Number: "))
    num2 = float(input("Enter Second Number: "))

    result = num1 / num2

    print("Result:", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Invalid input. Please enter numbers only.")
```

Read numbers from a file and calculate Total and Average. Handle the case if the file does not exist.

```python
try:
    file = open("numbers.txt", "r")

    numbers = []

    for line in file:
        numbers.append(int(line.strip()))

    total = sum(numbers)
    average = total / len(numbers)

    print("Total =", total)
    print("Average =", average)

    file.close()

except FileNotFoundError:
    print("Error: File does not exist.")
```
